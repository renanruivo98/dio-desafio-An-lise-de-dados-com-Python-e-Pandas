```python
#importando a biblioteca pandas
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/content/drive/MyDrive/datasets/Gapminder.csv",error_bad_lines=False, sep=";")
```

```python
df.head()
```

```python
df = df.rename(columns={"country":"Pais", "continent": "continente", "year":"Ano", "lifeExp":"Expectativa de vida", "pop":"Pop Total", "gdpPercap": "PIB"})
```

```python
#Alterando o estilo
plt.style.use("ggplot")
```

```python
df.groupby("Ano")["Expectativa de vida"].mean()
```

```python
df.groupby("Ano")["Expectativa de vida"].mean().plot(title = "Espectativa de vida x anos")
plt.xlabel("Anos")
plt.ylabel("Espectativa de vida")
plt.legend();
```

```python
df.groupby("Ano")["PIB"].mean().plot(title = "Media do PIB x anos")
plt.xlabel("Anos")
plt.ylabel("PIB")
plt.legend();
```

```python
Americas = df.loc[df["continente"] == "Americas"]
Americas.head()
```
